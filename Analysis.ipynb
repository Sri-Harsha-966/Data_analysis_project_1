```python
# Import Pandas.
import pandas as pd
```

```python
pd.set_option('display.max_columns', None) # So we can see all the colums.
df = pd.read_csv('Crime_Data_from_2020_to_Present.csv') # Loading data in to a Data Frame.
df.head() # Display a small section of data. 
```

```python
df.columns # Display all the column names as an array.
```

```python
df.info() # Shows us the type of data in each column and the number of non empty values.
df.isnull().sum() # Shows us how many NAN objects there are in each column.
```

```python
df['Crm Cd Desc'].nunique() # Displays the number of unique values in the column.
df['Crm Cd Desc'].value_counts() # Shows us the frequency of each value in a column.
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], format="%m/%d/%Y %I:%M:%S %p") # Convert the Date column to Pandas Datetime object.

# Extract Year, Month and day for easier time based sorting. 
df['YEAR_OCC'] = df['DATE OCC'].dt.year  # Extract year.
df['MONTH_OCC'] = df['DATE OCC'].dt.month  # Extract month.
df['DAY_OCC'] = df['DATE OCC'].dt.day  # Extract day of month.
df['DAY_OF_WEEK_OCC'] = df['DATE OCC'].dt.day_name()  # Extract weekday.

df[['DATE OCC', 'YEAR_OCC', 'MONTH_OCC', 'DAY_OCC', 'DAY_OF_WEEK_OCC']].head() # Check the newly created columns.
```

```python
# Extracting Hour from TIME OCC columns.
df['TIME OCC'] = df['TIME OCC'].astype(str).str.zfill(4) # Converting TIME OCC int column to strings.
df['HOUR'] = df['TIME OCC'].str[:2].astype(int) # Slicing these strings gives us hours.

df[['TIME OCC', 'HOUR']].head() # Check the newly created hour columns.
```

```python
# Dropping unneccesary columns.
columns_to_drop = ['DR_NO','Date Rptd','AREA','AREA NAME','Rpt Dist No','Part 1-2','Crm Cd','Mocodes','Vict Sex','Vict Descent','Premis Cd','Premis Desc','Weapon Used Cd', 'Weapon Desc','Status','Status Desc','Crm Cd 1',
       'Crm Cd 2', 'Crm Cd 3', 'Crm Cd 4', 'LOCATION', 'Cross Street'] # The columns to be dropped are made in to an array.
df.drop(columns=columns_to_drop, inplace=True) # The columns in the array are dropped.
```

```python
df.info() # Checking the dataset to look for missing values and wrong data types.
```

```python
df.to_csv('Cleaned_Crime_Data.csv',index=False) # Converting the cleaned dataset back into a csv.
```
